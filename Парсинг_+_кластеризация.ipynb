Поиск специализаций по направлениям (категориям)

```python
def get_professional_roles():
    url = "https://api.hh.ru/professional_roles"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        for category in data['categories']:
            print(f"Category ID: {category['id']}, Name: {category['name']}")
            for role in category['roles']:
                print(f"  Role ID: {role['id']}, Name: {role['name']}")
    else:
        print(f"Request failed with status code: {response.status_code}")

get_professional_roles()
```

Получение id специализаций в конкретной категории

```python
def get_roles_from_category(category_id):
    url = "https://api.hh.ru/professional_roles"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        for category in data['categories']:
            if category['id'] == str(category_id):
                return [role['id'] for role in category['roles']]
        print(f"No roles found for category ID: {category_id}")
        return []
    else:
        print(f"Request failed with status code: {response.status_code}")
        return None
```

Сбор информации о вакансиях

```python
def get_vacancies(role_id):
    url = "https://api.hh.ru/vacancies"
    full_vacancy_list = []
    page = 0

    while True:
      params = {
          "per_page": 100,  # Количество вакансий на странице
          "page": page, # Страница
          "professional_role": role_id,  # ID профессиональной роли
          "experience": "noExperience", # без опыта, затем - "between1And3",
      }
      headers = {
          "User-Agent": "Your User Agent",
      }

      response = requests.get(url, params=params, headers=headers)

      if response.status_code == 200:
          data = response.json()
          # print(f"Всего вакансий: {data['found']}")
          vacancies = data.get("items", [])

          for vacancy in vacancies:
              vacancy_id = vacancy.get("id")
              vacancy_title = vacancy.get("name")
              vacancy_url = vacancy.get("alternate_url")
              # Получаем дополнительные данные о вакансии
              response_detail = requests.get(f"https://api.hh.ru/vacancies/{vacancy_id}")
              if response_detail.status_code == 200:
                  vacancy_data = response_detail.json()
                  vacancy_info = {
                        "title": vacancy_data.get("name"),
                        "description": vacancy_data.get("description"),
                        "skills": [skill.get("name") for skill in vacancy_data.get("key_skills", [])],
                        "url": vacancy_url,
                        "id": vacancy_id,
                        }
                  full_vacancy_list.append(vacancy_info)
              else:
                print(f"Не удалось получить данные о вакансии с ID {vacancy_id}")
      else:
        print(f"Request failed with status code: {response.status_code}")
        break
      page += 1

    df = pd.DataFrame(full_vacancy_list)
    return df
```

```python
def main():
    role_ids = get_roles_from_category(11)  # Получаем все проф роли для категории 11 - Информационные технологии
    if role_ids is not None:
        all_vacancies = pd.DataFrame()  # Объединяем все полученные вакансии в один DataFrame
        for role_id in role_ids:
            df_vacancies = get_vacancies(role_id)
            all_vacancies = pd.concat([all_vacancies, df_vacancies], ignore_index=True)

        # Сохраняем все вакансии в csv файл
        all_vacancies.to_csv("vacancies_new.csv", index=False)

if __name__ == "__main__":
    main()
```

```python
# Приводим к нижнему регистру и убираем лишние символы
def normalize_text(text):
    text = text.lower()
    text = re.sub(r'[:;,.!?]', '', text)
    return text.strip()
```

```python
def extract_strong_text(df):
    result_df = df.copy()
    for index, row in df.iterrows():
        # Работаем с HTML из описания
        soup = BeautifulSoup(row['description'], 'html.parser')
        strong_tags = soup.find_all('strong')
        strong_texts = {}
        for tag in strong_tags:
            strong_text = tag.get_text(strip=True)
            normalized_strong_text = normalize_text(strong_text)  # Нормализуем текст

            parent_element = tag.find_parent()
            text_after_strong = ''

            if parent_element:
                for sibling in tag.find_all_next():
                    if sibling.name == 'strong':
                        break

                    # Защита от неверного перехода, если дошли до другого уровня (например, <ul>)
                    if sibling.name in ['ul', 'p']:
                        if sibling.name == 'ul':
                            text_after_strong += str(sibling)
                        break

                    text_after_strong += str(sibling)

                # Очищаем текст
                text_after_strong = text_after_strong.strip()
                strong_texts[normalized_strong_text] = text_after_strong

        # Добавляем новые столбцы в DataFrame
        for strong_text, value in strong_texts.items():
            column_name = strong_text[:30]
            if column_name not in result_df.columns:
                result_df[column_name] = None
            result_df.at[index, column_name] = value

    # Убираем столбцы, которые содержат менее двадцати ненулевых значений
    result_df = result_df.loc[:, result_df.notna().sum() > 20]
    return result_df
```

```python
def clean_column_names(column_names, stop_words):
    cleaned_names = []
    for name in column_names:
        words = name.lower().split()  # Разделяем на слова и приводим к нижнему регистру
        words = [word for word in words if word not in stop_words]  # Удаляем стоп-слова
        cleaned_name = ' '.join(words)  # Объединяем оставшиеся слова
        cleaned_name = lemmatize_text(cleaned_name)  # Применяем лемматизацию
        cleaned_names.append(cleaned_name)
    return cleaned_names

result_df.columns = clean_column_names(result_df.columns, stop_words)
result_df_columns = result_df.columns

print(result_df.columns)
print(len(result_df_columns))
```

```python
def auto_group_columns_by_words(columns):
    groups = defaultdict(list)
    for col in columns:
        words = set(col.lower().split())  # Разделяем на слова и приводим к нижнему регистру
        found_key = None
        # Проверяем существующие группы на наличие пересечений
        for key in groups.keys():
            if any(words & set(k.lower().split()) for k in groups[key]):  # Если есть пересечение
                found_key = key
                break
        if found_key is not None:
            groups[found_key].append(col)  # Добавляем к найденной группе
        else:
            groups[col] = [col]  # Создаем новую группу
    return dict(groups)  # Преобразуем в словарь

auto_group_columns_by_words(result_df_columns)
```

```python
# Функция для очистки текста
def clean_text(text):
    if pd.isna(text):
        return ' '
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^а-яА-ЯёЁ\s]', '', text)
    text = text.lower()
    return text

# Инициализация лемматизатора
lemmatizer = WordNetLemmatizer()

# Функция для лемматизации
def lemmatize_text(text):
    words = nltk.word_tokenize(text)
    lemmatized = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized)
```

# Новый раздел

Для признаков: название, описание и ключевые навыки ('title', 'description', 'skill')

```python
text_columns = ['title', 'description', 'skill']
text_data = df_copy[text_columns].fillna('').agg(' '.join, axis=1)
```

```python
# 1. Векторизация с помощью TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words=stop_words)
tfidf_matrix = tfidf_vectorizer.fit_transform(text_data)
```

```python
# 2. Определение оптимального количества кластеров для KMeans с помощью метода локтя
inertia = []
silhouette_scores = []
k_values = range(3, 15)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tfidf_matrix)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(tfidf_matrix, kmeans.labels_))

# Визуализация метода локтя
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.title('Метод локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция')

plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker='o')
plt.title('Силуэтный анализ')
plt.xlabel('Количество кластеров')
plt.ylabel('Силуэтный коэффициент')

plt.show()
```

```python
# KMeans
best_kmeans_score = -1
best_kmeans_n_clusters = 0
for n_clusters in range(3, 15):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans_labels = kmeans.fit_predict(tfidf_matrix)
    silhouette_avg = silhouette_score(tfidf_matrix, kmeans_labels)
    if silhouette_avg > best_kmeans_score:
        best_kmeans_score = silhouette_avg
        best_kmeans_n_clusters = n_clusters
print(f"KMeans: Лучшее количество кластеров: {best_kmeans_n_clusters}, Silhouette Score: {best_kmeans_score:.4f}")

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(tfidf_matrix)
if len(set(dbscan_labels)) > 1:
    dbscan_silhouette = silhouette_score(tfidf_matrix, dbscan_labels)
    print(f"DBSCAN: Silhouette Score: {dbscan_silhouette:.4f}")
else:
    print("DBSCAN: Обнаружен только один кластер.")

# Спектральная кластеризация
best_spectral_score = -1
best_spectral_n_clusters = 0
for n_clusters in range(3, 15):
    spectral = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
    spectral_labels = spectral.fit_predict(tfidf_matrix)
    silhouette_avg = silhouette_score(tfidf_matrix, spectral_labels)
    if silhouette_avg > best_spectral_score:
        best_spectral_score = silhouette_avg
        best_spectral_n_clusters = n_clusters
print(f"Спектральная кластеризация: Лучшее количество кластеров: {best_spectral_n_clusters}, Silhouette Score: {best_spectral_score:.4f}")

# Birch
birch = Birch(n_clusters=None, threshold=0.5, branching_factor=50)
birch_labels = birch.fit_predict(tfidf_matrix)
if len(set(birch_labels)) > 1:
    birch_silhouette = silhouette_score(tfidf_matrix, birch_labels)
    print(f"Birch: Silhouette Score: {birch_silhouette:.4f}")
else:
    print("Birch: Обнаружен только один кластер.")
```

```python
n_cluster = 8
```

```python
# 3. Кластеризация с помощью KMeans
kmeans = KMeans(n_clusters=n_cluster, random_state=42)
kmeans_labels = kmeans.fit_predict(tfidf_matrix)

silhouette_kmeans = silhouette_score(tfidf_matrix, kmeans_labels)
print(f'Silhouette Score (KMeans, TF-IDF): {silhouette_kmeans}')
```

```python
# Визуализация результатов KMeans с использованием PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(tfidf_matrix.toarray())
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, cmap='rainbow')
plt.title('Кластеры вакансий (KMeans)')
plt.xlabel('PCA Компонент 1')
plt.ylabel('PCA Компонент 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# 4. DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(tfidf_matrix)
# Силаэт не может быть рассчитан, если есть меньше двух кластеров
if len(set(dbscan_labels)) > 1:
    silhouette_dbscan = silhouette_score(tfidf_matrix, dbscan_labels)
    print(f'Silhouette Score (DBSCAN): {silhouette_dbscan}')
else:
    print('DBSCAN найдено менее 2 кластеров, силуэт не может быть рассчитан.')
```

```python
# Визуализация результатов DBSCAN
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan_labels, cmap='rainbow')
plt.title('Кластеры вакансий (DBSCAN)')
plt.xlabel('PCA Компонент 1')
plt.ylabel('PCA Компонент 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# 5. Спектральная кластеризация
spectral = SpectralClustering(n_clusters=n_cluster, assign_labels='discretize', random_state=42)
spectral_labels = spectral.fit_predict(tfidf_matrix)
silhouette_spectral = silhouette_score(tfidf_matrix, spectral_labels)
print(f'Silhouette Score (Spectral Clustering): {silhouette_spectral}')
```

```python
# Визуализация результатов спектральной кластеризации
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=spectral_labels, cmap='rainbow')
plt.title('Кластеры вакансий (Спектральная кластеризация)')
plt.xlabel('PCA Компонент 1')
plt.ylabel('PCA Компонент 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# 6. Birch кластеризация
birch = Birch(n_clusters=n_cluster)
birch_labels = birch.fit_predict(tfidf_matrix)
silhouette_birch = silhouette_score(tfidf_matrix, birch_labels)
print(f'Silhouette Score (Birch Clustering): {silhouette_birch}')
```

```python
# Визуализация результатов Birch
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=birch_labels, cmap='rainbow')
plt.title('Кластеры вакансий (Birch кластеризация)')
plt.xlabel('PCA Компонент 1')
plt.ylabel('PCA Компонент 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# 2. Определение оптимального количества кластеров для KMeans с помощью метода локтя
inertia = []
silhouette_scores = []
k_values = range(3, 15)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tfidf_matrix)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(tfidf_matrix, kmeans.labels_))

# Визуализация метода локтя
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.title('Метод локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция')

plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker='o')
plt.title('Силуэтный анализ')
plt.xlabel('Количество кластеров')
plt.ylabel('Силуэтный коэффициент')

plt.show()
```

```python
# KMeans
best_kmeans_score = -1
best_kmeans_n_clusters = 0
for n_clusters in range(3, 15):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans_labels = kmeans.fit_predict(tfidf_matrix)
    silhouette_avg = silhouette_score(tfidf_matrix, kmeans_labels)
    if silhouette_avg > best_kmeans_score:
        best_kmeans_score = silhouette_avg
        best_kmeans_n_clusters = n_clusters
print(f"KMeans: Лучшее количество кластеров: {best_kmeans_n_clusters}, Silhouette Score: {best_kmeans_score:.4f}")

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(tfidf_matrix)
if len(set(dbscan_labels)) > 1:
    dbscan_silhouette = silhouette_score(tfidf_matrix, dbscan_labels)
    print(f"DBSCAN: Silhouette Score: {dbscan_silhouette:.4f}")
else:
    print("DBSCAN: Обнаружен только один кластер.")

# Спектральная кластеризация
best_spectral_score = -1
best_spectral_n_clusters = 0
for n_clusters in range(3, 15):
    spectral = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
    spectral_labels = spectral.fit_predict(tfidf_matrix)
    silhouette_avg = silhouette_score(tfidf_matrix, spectral_labels)
    if silhouette_avg > best_spectral_score:
        best_spectral_score = silhouette_avg
        best_spectral_n_clusters = n_clusters
print(f"Спектральная кластеризация: Лучшее количество кластеров: {best_spectral_n_clusters}, Silhouette Score: {best_spectral_score:.4f}")

# Birch
birch = Birch(n_clusters=None, threshold=0.5, branching_factor=50)
birch_labels = birch.fit_predict(tfidf_matrix)
if len(set(birch_labels)) > 1:
    birch_silhouette = silhouette_score(tfidf_matrix, birch_labels)
    print(f"Birch: Silhouette Score: {birch_silhouette:.4f}")
else:
    print("Birch: Обнаружен только один кластер.")
```

# BERT

```python
# Загружаем токенизатор
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Загружаем модель
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
df_copy = pd.read_csv('vacancies_copy.csv',nrows=500)
text_columns = ['skills', 'требования', 'обязанности', 'условия', 'важно', 'плюсом',
                'задачи', 'ожидаем', 'наши ожидания']
text_data = df_copy[text_columns].fillna('').agg(' '.join, axis=1)
```

```python
text_list = text_data.tolist()
```

```python
# Токенизация текстов
inputs = tokenizer(text_list, return_tensors='pt', padding=True, truncation=True)
```

```python
# Пропуск через модель (перегруз)
with torch.no_grad():
    outputs = model(**inputs)
```

```python
# Получение скрытых состояний для каждой строки
last_hidden_states = outputs.last_hidden_state
for i in range(len(text_list)):
    cls_embedding = last_hidden_states[i][0]
    print(f"Embedding для строки {i+1}: {cls_embedding}")
```
